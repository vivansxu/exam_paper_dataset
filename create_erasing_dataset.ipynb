## Label positions for pasting handwrittens

### Edit following:

- work_path: path of original images
- save_path: path for saving

### Usage:

- Double click the positions to paste handwritten images
- Press 'q' for next image
- A label.csv file would be saved to save_path

__Format of label.csv file__

name|pos_num|pos| | | 
:--:|:-----:|:-:|:-:|:-:
example.png|2|(100,200)|(150,250)| | 
apple.png|4|(101,201)|(50,50)|(1,2)|(3,4)

```python
work_path = '/Users/vivan/Downloads/exampaper/test'
save_path = '/Users/vivan/Downloads/exampaper'
```

```python
import os
import cv2
import numpy as np
import csv

image_names = os.listdir(work_path)

def record(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDBLCLK:
        position = '(' + str(x) + ',' + str(y) + ')'
        row.append(position)

wFile = open(os.path.join(save_path, 'label.csv'), "w", encoding = 'utf-8')
writer = csv.writer(wFile)
fileheader = ['name', 'pos_num', 'pos']
writer.writerow(fileheader)

for image_name in image_names:
    if not image_name.endswith('.png'):
        continue
    
    img = cv2.imread(os.path.join(work_path, image_name))
    
    
    row = []
    row.append(image_name)

    cv2.namedWindow(image_name) 
    cv2.setMouseCallback(image_name,record)
    
    while(1):
        cv2.imshow(image_name,img)
        if cv2.waitKey(100) == ord('q'):
            break
    
    cv2.destroyAllWindows()  
    
    pos_num = len(row) - 1
    row.insert(1, pos_num)
    writer.writerow(row)
    
wFile.close()
            
cv2.destroyAllWindows()

print('Finished! File saved at ' + save_path + '/label.csv')
```
